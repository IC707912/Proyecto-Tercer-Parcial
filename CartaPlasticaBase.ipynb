```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_file = 'CartaPlastica.csv'
data = pd.read_csv(data_file, header=None)

X = data.iloc[:,0:2].values
Y = data.iloc[:,2].values
```

```python
plt.figure(figsize = (8,6))
plt.scatter(X[:,0], X[:,1], c=Y)
plt.show()
```
